## Exercise 1: Clean and Format the Data

While the dataset has been partially cleaned and formatted, we still need to perform several data cleaning tasks. Specifically, we need to standardize the column names, clean null values, convert the effective_to_date column to datetime, and extract the months from the dataset and store them in a separate column. 

To accomplish these tasks, we will use the functions created in the previous step to standardize the column names and deal with null values, and then we will apply additional functions to convert the effective_to_date column to datetime and extract the months.

Save the clean dataset into a csv file.

```python
marketing_customer_analysis = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
marketing_customer_analysis
```

```python
# Standardizing the column names
print("The main cleaning function doesn't work with this data frame because it already has the 'customer lifetime value' and 'number of open complaints' cleaned without symbols, so it raises an error. That's why in this case we have to apply the individual functions to clean and format it.")
marketing_customer_analysis = cleaning_functions.standardize_column_names(marketing_customer_analysis)
marketing_customer_analysis
```

```python
# Checking the rows in which there are null values
marketing_customer_analysis.isnull().sum()
```

```python
# Cleaning null values
marketing_customer_analysis = cleaning_functions.drop_row_all_null(marketing_customer_analysis)
marketing_customer_analysis = cleaning_functions.replace_null_categorical(marketing_customer_analysis,'state')
marketing_customer_analysis = cleaning_functions.replace_null_categorical(marketing_customer_analysis,'response')
marketing_customer_analysis = cleaning_functions.replace_null_categorical(marketing_customer_analysis,'vehicle_class')
marketing_customer_analysis = cleaning_functions.replace_null_categorical(marketing_customer_analysis,'vehicle_size')
marketing_customer_analysis = cleaning_functions.replace_null_categorical(marketing_customer_analysis,'vehicle_type')
marketing_customer_analysis = cleaning_functions.replace_null_numerical_median(marketing_customer_analysis,'months_since_last_claim')
marketing_customer_analysis = cleaning_functions.replace_null_numerical_median(marketing_customer_analysis,'number_of_open_complaints')
marketing_customer_analysis
```

```python
# Checking if the null values have been properly cleaned
marketing_customer_analysis.isnull().sum()
```

```python
# Converting the effective_to_date column to datetime
marketing_customer_analysis['effective_to_date'] = pd.to_datetime(marketing_customer_analysis['effective_to_date'], errors='coerce')
marketing_customer_analysis['effective_to_date']
```

```python
# Checking the effective_to_date column's type has been properly changed
marketing_customer_analysis.dtypes
```

```python
# Extracting the months from the dataset and storing them in a separate column
months=[]
for i in marketing_customer_analysis['effective_to_date'].keys():
    months.append(marketing_customer_analysis['effective_to_date'][i].month)
marketing_customer_analysis['months']=months
marketing_customer_analysis
```

```python
# Saving the clean dataset to a new CSV file
marketing_customer_analysis.to_csv('clean_marketing_customer_analysis.csv',index = False)
```

## Exercise 2: Structuring Data

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

3. You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table. 

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# 1.Summary table showing the total revenue for each sales channel 
display(marketing_customer_analysis.pivot_table(values=['customer_lifetime_value'],index=['sales_channel'],aggfunc={'customer_lifetime_value':'sum'}).round(decimals=2))
print("As in the exercise statement it isn't specified the time period they want, I used the column customer_lifetime_value as total revenue because of its description: estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company.")
print("Looking at this table we can see that the sales channel that brought the most sales in terms of total revenue is Agent, while the channel that brought the least is Web.")
```

```python
# 2. Pivot table that shows the average customer lifetime value per gender and education level
display(marketing_customer_analysis.pivot_table(values=['customer_lifetime_value'],index=['gender','education']))
print("Looking at this table we can see that, for females, the highest average customer lifetime value belongs to the individuals with an educational level of high school or below. In the male case, it belongs to the Master category, but is closeliy followed by high school or below. The lowest average customer lifetime value for females belongs to the Doctor category, as well as for males.")
```

```python
# 3. Summary table showing the number of complaints by policy type and month
complaints_policy_month=marketing_customer_analysis.groupby(['policy_type','months'])[['number_of_open_complaints']].sum().reset_index()
complaints_policy_month.pivot(index='policy_type',columns='months')
```
